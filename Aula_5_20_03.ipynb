:#**OPERADORES ARITMÉTICOS DIVIDIDAS JUNTAS**


```python
#Entrada 1

print ("----OPERADORES ARITMÉTICOS----")
print ("                                                     ")

print ("SOMA SIMPLES:")
print ("                                                     ")

a = float (input ("digite o primeiro valor: "))
b = float (input ("digite o segundo valor: "))
print ("                                                     ")

#Processamento

soma = a+b

#Saída

print (f"O resultado da soma de {a} mais {b} é igual: {soma}" )
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")







#Entrada 2
print ("-----SUBTRAÇÃO-----")
print ("                                                     ")

c = float (input ("digite o primeiro valor: "))
d = float (input ("digite o segundo valor: "))
print ("                                                     ")

#Processamento

sub = c-d

#Saída

print (f"O resultado da soma de {c} menos {d} é igual: {sub}" )
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")







#Entrada 3
print ("-----MULTIPLICAÇÃO-----")
print ("                                                     ")

e = float (input ("digite o primeiro valor: "))
f = float (input ("digite o segundo valor: "))
print ("                                                     ")

#Processamento

mult = e*f

#Saída

print (f"O resultado da soma de {e} vezes {f} é igual: {mult}" )
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")








#Entrada 4
print ("-----DIVISÃO-----")
print ("                                                     ")

g = float (input ("digite o primeiro valor: "))
h = float (input ("digite o segundo valor: "))
print ("                                                     ")

#Processamento

divi = g/h

#Saída

print (f"O resultado da soma de {g} dividido por {h} é igual: {divi}" )
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")









#Entrada 5
print ("-----MÉDIA-----")
print ("                                                     ")

i = float (input ("digite o primeiro valor: "))
j = float (input ("digite o segundo valor: "))
k = float (input ("digite o terceiro valor: "))

print ("                                                     ")

#Processamento

media = i + j + k

resultado = media / 3

#Saída

print (f" o resultado da média dos 3 valores é : {resultado}")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")










#Entrada 6
print ("-----DIVISÃO INTEIRA E RESTO-----")
print ("                                        ")

l = float (input ("digite o valor: "))


print ("                                                     ")

#Processamento

dobro = l*2
metade = l/2
resto = l%3

#Saída

print (f" o resultado do dobro de {l} é : {dobro}")
print ("                                        ")
print (f" o resultado da metade de {l} é : {metade}")
print ("                                        ")
print (f" o resultado do resto de {l} é : {resto}")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")








#Entrada 7

print ("------CONVERSOR DE GRAUS EM FAHRENHEIT------")
print ("                                            ")
print ("TEMPERATURA EM GRAUS:                       ")
print ("                                            ")
graus = float (input ("digite a temperatura em graus: "))

#Processamento

fahrenheit = graus*(9/5)+32

#Saída
print ("                                            ")
print ("CONVERSÃO:                                  ")
print ("                                            ")
print (f"A temperatura em fahrenheit é: {fahrenheit} ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")








#Entrada 8

print ("-----CÁLCULO DE SALÁRIO-----")
print ("                                        ")

m = float (input ("digite o salário: "))


print ("                                                     ")

#Processamento

salario = 10/100 * m/1
novo_salar = salario + m

#Saída

print (f" O resultado dos 10% de {m} é : {salario}")

print (f" O novo salário do funcionário e de: {novo_salar}")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")






#Entrada 9

print ("-----ÁREA DE UM RETÂNGULO-----")
print ("                                        ")

n= float (input ("digite a base: "))
o= float (input ("digite a altura: "))

# Processamento

área = n * o

# Saída

print (f" A área do retângulo é : {área}")

print ("                                                     ")
print ("                                                     ")
print ("                                                     ")
print ("                                                     ")








#Entrada 10

print ("-----CONSUMO DE COMBUSTIVEL-----")
print ("                                        ")

p= float (input ("Digite a distância percorrida (km): "))
q= float (input ("Digite o combustível gasto (litros): "))

# Processamento

consumo_medio = p / q

# Saída

print (f" O consumo médio é: {consumo_medio}")
```
