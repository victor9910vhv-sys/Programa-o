```python
# Escrever um programa que leia as notas dos quatro bimestre, nome do aluno e calcule a média.

#média menor que 5 - reprovado
#média menor que 7 - recuperação
#média maior que 7 - aprovado



nome = str (input("Digite seu nome: "))

nota1 = float (input ("Digite a 1º nota do bimestre: "))
nota2 = float (input ("Digite a 2º nota do bimestre: "))
nota3 = float (input ("Digite a 3º nota do bimestre: "))
nota4 = float (input ("Digite a 4º nota do bimestre: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

if media < 5:
  print("Situação: Reprovado")
elif media < 7:
  print("Situação: Recuperação")
else:
  print("Situação: Aprovado")
```
