# **Atividade 2**

Escrever um algoritimo e o programa que receba 4 notas do aluno e calcule a média dessas notas e mostre a média

# Entrada

Receber 4 números (Notas)

# Processamento

Calcular a média dos números, somando os números e depois dividindo por 4

# Saída

Mostrar a média

```python
# Entrada

nota1= 10
nota2= 10
nota3= 10
nota4= 10

# Processamento

soma = nota1 + nota2 + nota3 + nota4

resultado = soma / 4

# Saída

print (f" o resultado da média do aluno é : {resultado}")
```
