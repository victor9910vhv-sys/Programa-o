```python
#QUESTÃO 02 – Soma dos Números

numero = int(input(" Digite um numero: "))
soma = 0

for i in range (numero + 1 ):
  soma += i
  print(soma)
```

```python
#QUESTÃO 03 – Tabuada

numero = int(input(" Digite um numero: "))

for i in range (0, 11):
  mult = numero * i
  print(f"{numero} x {i} = {mult}")
```

```python
#QUESTÃO 05 – Caixa Eletrônico

v = int(input("Digite o valor do saque: "))

nota = 100

while v > 0:

    qtd = v // nota

    if qtd > 0:
        print(f"{qtd} nota(s) de R${nota}")
        v = v % nota

    if nota == 100:
        nota = 50
    elif nota == 50:
        nota = 20
    elif nota == 20:
        nota = 10
    elif nota == 10:
        nota = 5
    elif nota == 5:
        nota = 2
    elif nota == 2:
        nota = 1

```
