```python
# 1.Par ou Impar

numero = int(input("Digite um número: "))

if numero % 2 == 0:
  print(f"O número {numero} é par.")
else:
  print(f"O número {numero} é ímpar.")
```

```python
# 2.Comparação entre dois números

numero1 = int (input ("digite o primeiro número: "))

numero2 = int (input ("digite um segundo número: "))

if numero1 > numero2:
  print (f"{numero1} é maior que {numero2}")

else:
  print (f"{numero1} é menor que {numero2}")
```

```python
# 4.Classificação de idade

idade = int(input("Digite sua idade: "))


if idade <= 12:
  print("Você é uma criança")

elif idade >= 13 and idade <= 17:
  print("Você é um adolescente")

elif idade >= 18 and idade <= 59:
  print("Você é um adulto")

else:
  print("Você é um idoso")
```

```python
# 6.Positivo, negativo ou zero

num = int (input("Digite um número: "))

if num > 0:
  print ("Seu número e positivo")

elif num < 0:
  print ("Seu número e negativo")

else:
   print ("Seu número e zero")
```

```python
# 7.Calculadora simples

print ("|------------CALCULADORA------------|")
print ("|                                   |")
print ("|---------------MENU----------------|")
print ("|Opção = 1 = Soma                   |")
print ("|Opção = 2 = Subtração              |")
print ("|Opção = 3 = Multiplicação          |")
print ("|Opção = 4 = Divisão                |")
print ("|-----------------------------------|")

# Entrada
numero1 = float (input ("Digite um número: "))
opcao  = int (input("Escolha uma das opções: "))
numero2 = float (input("Digite outro número: "))

# Processamento / saída

if opcao == 1:
  print (f" A soma de {numero1} e {numero2} é igual a {numero1 + numero2} :")

elif opcao == 2:
  print (f" A subtração de {numero1} e {numero2} é igual a {numero1 - numero2} :")

elif opcao == 3:
  print (f" A multiplicação de {numero1} e {numero2} é igual a {numero1 * numero2} :")

elif opcao == 4:
  print (f" A divisão de {numero1} e {numero2} é igual a {numero1 : numero2} :")

else:
  print ("Opção Inválida")
```

```python
# 10.Desconto em compras

valor = float(input("qual o valor de sua compra? R$"))

if valor >= 100:
  desc = valor * 0.1
  prec = valor - desc
  print(f"sua compra ganhou 10% de desconto e ficou {prec}")

elif valor >= 50 and valor < 100:
  desc = valor * 0.05
  prec = valor - desc
  print(f"sua compra ganhou 5% de desconto e ficou {prec}")

else:
  print(f"sua compra não ganhou desconto, e ficou {valor}")
```
