```python
# 1 - Contagem simples, o usuario digita um número e o programa imprimi do zero até o número.

numero = int(float(input(" Digite um numero: ")))

for i in range (numero + 1):
  print(i)
```

```python
# 2 - Contagem regressiva, o usuario digita um número e o programa imprimi do número até zero.

n = int(float(input(" Digite um numero: ")))

for i in range (0, n + 1):
  print (n - i)
```

```python
# 3 - Números pares, o usuário digita um número e o programa imprimi os números pares.

numero = int(input(" Digite um numero: "))

for i in range (numero + 2):
  if i % 2 == 0:
    print (i)
```

```python
# 4 - Números impares, o usuário digita um número e o programa imprimi os números impares.

numero = int(input(" Digite um numero: "))

for i in range (numero + 2):
  if i % 2 == 1:
    print (i)
```

```python
# 5 - Soma, o usuario digita um número e o programa soma os números do zero até o número.

numero = int(input(" Digite um numero: "))
soma = 0

for i in range (numero + 1):
  soma += i
  print(soma)
```

```python
# 6 - Tabuada, o usuário digita um número e o programa imprimi a tabuada do número.

numero = int(input(" Digite um numero: "))

for i in range (0, 11):
  mult = numero * i
  print(f"{numero} x {i} = {mult}")
```

```python
# 7 - Contar letras, o usuário digita uma palavra e o programa conta as letras da palavra.

palv = input(" Digite uma palavra: ")
contg = len(palv)

print(f"A palavra {palv} tem {contg} letras.")
```

```python
# 8 - Contar vogais, o usuário digita uma palavra e o programa conta as vogais.

palavra = input(" Digite uma palavra: ")
vogais = "aeiouAEIOU"
vogal = 0

for letra in palavra:
  if letra in vogais:
    vogal += 1

print(f"A palavra {palavra} tem {vogal} vogais.")
```

```python
# 9 - Maior número digitado, o usuário digita quatro números e o programa imprimi o maior número digitado.

num1 = int(input("Digite o 1º: "))
num2 = int(input("Digite o 2º: "))
num3 = int(input("Digite o 3º: "))
num4 = int(input("Digite o 4º: "))

maior = max(num1, num2, num3, num4)

print(f"O maior número digitado é: {maior}")
```

```python
# 10 - O usuário digita um número entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5.

numero = int(input("Digite um número entre 1 e 100: "))

print(f"Números de 1 a {numero} divisíveis por 3 e 5:")
for i in range(1, numero + 1):
  if i % 3 == 0 and i % 5 == 0:
    print(i)
```
